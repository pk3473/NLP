### SEQ2SEQ Models

    The main idea is to learn the probability of occurrence of an element based on previous elements in the sequence.
    
    Examples:
    - Audio, Signals, Time Series
    - Text: Sequence of words

    Applications:
    - Machine Translation
    - Text Summarization
    - Text Generation
    - Conversational Agents (Chatbots)

## Recurrent Neural Networks - RNN's

    - Activation outputs propagates only in one direction incase of Feed-Forward Direction where as incase of RNN it propagates in both directions (from input to output and output to input)
    
    - It essentially creates loops in the network and up on unfold we get similar to Neural Network    
    - Loops in the network act as memory state of the neurons which allows the neurons to see and capture have been learned
    - RNN's are associated with a problem called Vanishing Gradient

### Vanishing Gradient

    After performing large number of epochs or iterations, Network tends to forget what was learned in the earlier layers.  This is because in the training process the model tries to optimize the weights & biases associated with the layer.     However in RNN, gradient becomes so small even close to zero which makes difficult to remember earlier learning.
      
    To overcome this problem, a new type of RNN is developed known as Long Short Term Memory (LSTM).

### Long Short Term Memory

    - LSTMs are a special type of RNNs with an additional state 'Cell State' apart from 'Memory State'
    - This is used to make adjusment in information flow. With Cell state it will remember learnings more selectively.
    - It predicts the likelihood of occurence of a word based on the previous sequence of words.

```python
from keras.models import Sequential 
from keras.layers import Dense, LSTM, Embedding
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical
import numpy as np 
```

```python
data = "rain rain go away come again another day little johny wants to play johny wants to play again in the rain"
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data]) # This data should be list of sentences. In this case, we have only one sentence
                               # Due to this every will have an index
```

```python
# Converting input data into sequences

sequence = tokenizer.texts_to_sequences([data])
sequence

# 1- rain, 7- go, 8- away

# you can see two lists below
```

```python
sequence = tokenizer.texts_to_sequences([data])[0] # to get only one list
sequence 
```

```python
tokens = tokenizer.word_index
tokens
```

```python
tokens =len(tokenizer.word_index)+1
tokens
```

```python
# Preparing the data for text generation problem
# Creating Bigrams "X Y" ==> X - Independent Variable; Y - Target Variable

sequences =[]

for i in range(1, len(sequence)): # sequence is the one which obtained earlier
    seq = sequence [i-1: i+1]
    # print(seq)
    sequences.append(seq)
sequences = np.array(sequences)
```

```python
sequences
```

```python
train_x, train_y = sequences[:,0], sequences[:,1] #: ==> all the rows; 0 ==> first element; 1 ==> second element

# Convert target variable into categorical variable

#train_y = to_categorical(train_y, num_classes = tokens) # num_classes ==> Number of classes 

train_y = to_categorical(train_y,num_classes = tokens)
```

#### LSTM Architecture

    Input >> Embedding >> LSTM >> Multi classification

```python
model = Sequential()

model.add(Embedding(tokens,10, input_length = 1)) # First Layer - Embedding Layer

model.add(LSTM(32)) # Number of units = 32; this will perform LSTM Operation essentially which is to learn if a word is given 
                    # what is the likelihood of next word. This is a hyperparmeter. It can be changed and tuned.
    
model.add(Dense(tokens,activation = "softmax")) # Softmax is useful during multi classification problem

model.compile(loss='categorical_crossentropy',optimizer = 'adam',metrics = ["accuracy"]) 
                                                        # different optimizers can be tried out
                                                        # accuracy ==> Classification

model.summary()
```

```python
model.fit(train_x,train_y,epochs = 500,verbose = 1)
```

```python
# In every epoch we can see loss is getting decreased

# Now we can check how this model is performing for text generation or on new text
```

```python
# The idea of this function is given an input text what are the particular words can come after this text

def generate_text(input_text):
    result = input_text
    
    for j in range(4): # want to predict next four words
        
        seq = np.array(tokenizer.texts_to_sequences([input_text])[0]) #repeating the preprocessing steps
                                                            # accessing only first element intially and will convert this into
                                                            # numpy array. This is input sequence
        
        pred_seq = model.predict_classes(seq) # predicting sequence. This will return index
        
        for word, index in tokenizer.word_index.items():
            if index == pred_seq:
                out_word = word
                break
                
        input_text = out_word
        result += " "+out_word
    return result
```
